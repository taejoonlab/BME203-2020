```python
# Read a FASTA file
def read_fasta(tmp_filename):
    rv = dict()
    f = open(tmp_filename, 'r')
    for line in f:
        if line.startswith('>'):
            tmp_h = line.strip().lstrip('>')
            rv[tmp_h] = ''
        else:
            rv[tmp_h] += line.strip().replace(' ', '')
    f.close()
    return rv


# The function to get the reverse complementary sequences
rc = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
def revcomp(tmp_seq):
    # from the end of input sequence to the beginning, 
    # replace each nucleotide to its complementary one.
    #
    # The single line code at the end of this function is equivalent to the following 4 lines.
    # rv = []
    # for tmp_n in tmp_seq[::-1]:
    #     rv.append(rc[tmp_n])
    # return ''.join(rv)
    
    return ''.join([rc[x] for x in tmp_seq[::-1]])
    

# Read 2019nCoV primers from a FASTA file
filename_primers = '../data/2019nCoV_primers.fa'
primer_list = read_fasta(filename_primers)
for p_h, p_seq in primer_list.items():
    print(p_h, p_seq)
    
print('')
    
# Read 2019nCoV genomes from a FASTA file    
filename_genomes = '../data/2019nCoV_genomes.2020_02_03.fa'
genome_list = read_fasta(filename_genomes)
for g_h, g_seq in genome_list.items():
    print(g_h.split()[0], g_seq[:30])

print('')
#############################################################################################################    
# Calculate the length of PCR amplicons for Covid-19 diagnostics.
# Length of amplicon 
amp_len_N1 = []
amp_len_N2 = []
amp_len_N3 = []

amp_s_N1 = []
amp_s_N2 = []
amp_s_N3 = []

amp_p = []

# Find the position of primers (F, R) on each genome sequence
for g_h, g_seq in genome_list.items():
    for p_h, p_seq in primer_list.items():
        if "N1" in p_h:                                                     
            if p_h[-1] == 'F':                      
                N1_primer_posi_S = g_seq.find(p_seq)                                  # Find the position of each primers(N1/N2/N3) on each genome sequence
                
                amp_p.append(p_seq) 
                    
            if p_h[-1] == 'R':
                N1_CompRP = revcomp(p_seq)
                N1_primer_posi_E = g_seq.find(N1_CompRP)
            
                N1_Amp_length = abs(N1_primer_posi_S - N1_primer_posi_E) + len(N1_CompRP)  # Calculate the length of PCR amplicons 
                amp_len_N1.append(N1_Amp_length)
                
                N1_amp = g_seq[N1_primer_posi_S:N1_primer_posi_E] + N1_CompRP

                amp_s_N1.append(N1_amp)
                amp_p.append(N1_CompRP)   
                
        
        if "N2" in p_h:
            if p_h[-1] == 'F':
                N2_primer_posi_S = g_seq.find(p_seq)                                  
                
                amp_p.append(p_seq) 
                
            if p_h[-1] == 'R':
                N2_CompRP = revcomp(p_seq)
                N2_primer_posi_E = g_seq.find(N2_CompRP)
                
                
                    
                N2_Amp_length = abs(N2_primer_posi_S - N2_primer_posi_E) + len(N2_CompRP)
                amp_len_N2.append(N2_Amp_length)
            
                N2_amp = g_seq[N2_primer_posi_S:N2_primer_posi_E] + N2_CompRP

                amp_s_N2.append(N2_amp)
                amp_p.append(N2_CompRP)
        
        if "N3" in p_h:
            if p_h[-1] == 'F':
                N3_primer_posi_S = g_seq.find(p_seq)                                  
                
                amp_p.append(p_seq)
                
            if p_h[-1] == 'R':
                N3_CompRP = revcomp(p_seq)
                N3_primer_posi_E = g_seq.find(N3_CompRP)
                
            

                N3_Amp_length = abs(N3_primer_posi_S - N3_primer_posi_E) + len(N3_CompRP)
                amp_len_N3.append(N3_Amp_length)
                
                N3_amp = g_seq[N3_primer_posi_S:N3_primer_posi_E] + N3_CompRP
                
                
                amp_s_N3.append(N3_amp)
                amp_p.append(N3_CompRP)
                
dic_A_s_N1 = dict(zip(range(len(amp_s_N1)), amp_s_N1)) 
dic_A_s_N2 = dict(zip(range(len(amp_s_N2)), amp_s_N2))                 
dic_A_s_N3 = dict(zip(range(len(amp_s_N3)), amp_s_N3)) 

# 1. What is the length of amplicons generated by N1/N2/N3 primers? : each primers' mean of length 
N1_amp_M = sum(amp_len_N1)/len(amp_len_N1)
N2_amp_M = sum(amp_len_N2)/len(amp_len_N2)
N3_amp_M = sum(amp_len_N3)/len(amp_len_N3)





print("What is the length of amplicons generated by N1/N2/N3 primers?")

print("The mean of [2019-nCoV_N1] amplicon length: " + str(N1_amp_M) )
print("The mean of [2019-nCoV_N2] amplicon length: " + str(N2_amp_M) )
print("The mean of [2019-nCoV_N3] amplicon length: " + str(N3_amp_M) )

print('')
# 1-1. Are they all same among the genomes?

for k in amp_s_N1:
    n = 0
    amp_s_N1[n]
    #print(amp_s_N1[n])
    n = n+1
if amp_s_N1[n] == amp_s_N1[n+1]:
    print("yes,", "Amplicon seq: ", amp_s_N1[0])
else:
    print("no")
    
for j in amp_s_N2:
    n = 0
    amp_s_N2[n]
    #print("Amplicon seq: ",amp_s_N2[n])
    n = n+1
if amp_s_N2[n] == amp_s_N2[n+1]:
    print("yes,", "Amplicon seq: ", amp_s_N2[0])
else:
    print("no")
    
for t in amp_s_N3:
    n = 0
    amp_s_N3[n]
    #print("Amplicon seq: ",amp_s_N3[n])
    n = n+1
if amp_s_N3[n] == amp_s_N3[n+1] :
    print("yes,", "Amplicon seq: ", amp_s_N3[0])
else:
    print("no")
                
amp_seq_dict = dict(zip(['amp_s_N1','amp_s_N2','amp_s_N3'],[amp_s_N1[0],amp_s_N2[0],amp_s_N3[0]]))
#print(amp_seq_dict)






```
